```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

path = '/content/drive/MyDrive/Int Conference/Data Label/LPDP_cleaned_label.csv'

df = pd.read_csv(path)

print("Jumlah data:", df.shape)
df.head()
```

```python
sentiment_counts = df['label_voting'].value_counts()

plt.figure(figsize=(6,4))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values)

plt.title("Distribusi Sentimen Opini Publik terhadap LPDP")
plt.xlabel("Kategori Sentimen")
plt.ylabel("Jumlah Tweet")

plt.show()

print(sentiment_counts)
```

**Perbandingan Metode Labeling**

```python
methods = ['label_textblob','label_vader','label_senti_afinn']

for m in methods:
    print("\nDistribusi Sentimen -", m)
    print(df[m].value_counts())
```

**Konsistensi Antar Metode**

```python
agreement = (
    (df['label_textblob'] == df['label_vader']) &
    (df['label_vader'] == df['label_senti_afinn'])
)

agreement_rate = agreement.mean()*100

print("Tingkat agreement antar metode:", agreement_rate,"%")
```

**Kata yang Paling Sering Muncul**

```python
from collections import Counter

text = " ".join(df['cleaned'].astype(str))

word_freq = Counter(text.split())

top_words = word_freq.most_common(20)

top_df = pd.DataFrame(top_words, columns=['word','freq'])

plt.figure(figsize=(8,5))

sns.barplot(data=top_df, x='freq', y='word')

plt.title("20 Kata yang Paling Sering Muncul")

plt.show()
```

**Kata Dominan Berdasarkan Sentimen**

```python
positive_text = " ".join(df[df['label_voting']=="Positif"]['cleaned'].astype(str))
negative_text = " ".join(df[df['label_voting']=="Negatif"]['cleaned'].astype(str))

pos_words = Counter(positive_text.split()).most_common(15)
neg_words = Counter(negative_text.split()).most_common(15)

pos_df = pd.DataFrame(pos_words, columns=['word','freq'])
neg_df = pd.DataFrame(neg_words, columns=['word','freq'])

plt.figure(figsize=(8,5))
sns.barplot(data=pos_df, x='freq', y='word')
plt.title("Kata Dominan pada Sentimen Positif")
plt.show()

plt.figure(figsize=(8,5))
sns.barplot(data=neg_df, x='freq', y='word')
plt.title("Kata Dominan pada Sentimen Negatif")
plt.show()
```

```python
import gensim
from gensim import corpora
from gensim.models import LdaModel

# tokenisasi
texts = df['cleaned'].astype(str).apply(lambda x: x.split())

# dictionary
dictionary = corpora.Dictionary(texts)

# corpus
corpus = [dictionary.doc2bow(text) for text in texts]

# LDA model
lda_model = LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=5,
    passes=10,
    random_state=42
)

# tampilkan topik
topics = lda_model.print_topics(num_words=10)

for topic in topics:
    print(topic)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df['cleaned'].isna().sum()
```

```python
# ganti NaN jadi string kosong
df['cleaned'] = df['cleaned'].fillna("")

vectorizer = CountVectorizer(
    ngram_range=(2,2),
    max_features=20
)

X = vectorizer.fit_transform(df['cleaned'])

ngrams = vectorizer.get_feature_names_out()

counts = X.sum(axis=0).A1

ngram_df = pd.DataFrame({
    "ngram":ngrams,
    "freq":counts
}).sort_values("freq",ascending=False)

ngram_df
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=ngram_df, x="freq", y="ngram")

plt.title("Bigram yang Paling Sering Muncul")

plt.show()
```
