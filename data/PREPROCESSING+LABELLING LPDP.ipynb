Hapus Mention & Hashtag + Segmentasi Hashtag

```python
# @title
def clean_mentions_hashtags(text):
    # Hapus mention
    text = re.sub(r"@\w+", "", text)
    # Ganti hashtag jadi kata tanpa "#"
    text = re.sub(r"#(\w+)", r"\1", text)
    return text
```

```python
#id
from googletrans import Translator
translator = Translator()
def translate_to_english(text):
    if isinstance(text, str) and text.strip():
        try:
            return translator.translate(text, dest='en').text
        except Exception:
            return "" # Kembalikan string kosong jika gagal

    return ""
```

```python
LPDP_clean = pd.read_csv(
    '/content/drive/MyDrive/Int Conference/Data Clean/LPDP_partial_translation.csv'
)
```

```python
# LPDP_clean['english_translation'] = ""
```

```python
mask = LPDP_clean['english_translation'].isna() | (LPDP_clean['english_translation'] == "")

subset = LPDP_clean.loc[mask]
```

```python
from tqdm import tqdm
tqdm.pandas()

batch_size = 1000

for start in range(0, len(subset), batch_size):
    end = start + batch_size

    indices = subset.index[start:end]

    LPDP_clean.loc[indices, 'english_translation'] = \
        LPDP_clean.loc[indices, 'normalisasi_text'].progress_apply(translate_to_english)

    # AUTO SAVE tiap batch
    LPDP_clean.to_csv(
        '/content/drive/MyDrive/Int Conference/Data Clean/LPDP_partial_translation.csv',
        index=False
    )

    print(f"Saved batch {start} to {end}")
```

labelling-text blob

```python
!pip install textblob
```

```python
from textblob import TextBlob

def label_textblob(text):
    if isinstance(text, str):
        polarity = TextBlob(text).sentiment.polarity
        if polarity > 0:
            return "Positif"
        elif polarity < 0:
            return "Negatif"
        else:
            return "Netral"
    return "Netral"
```

```python
LPDP_clean['label_textblob'] = LPDP_clean['english_translation'].apply(label_textblob)
```

labelling-vader

```python
!pip install vaderSentiment
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

def label_vader(text):
    if isinstance(text, str):
        score = analyzer.polarity_scores(text)
        if score['compound'] >= 0.05:
            return "Positif"
        elif score['compound'] <= -0.05:
            return "Negatif"
        else:
            return "Netral"
    return "Netral"
```

```python
LPDP_clean['label_vader'] = LPDP_clean['english_translation'].apply(label_vader)
```

labelling-senti(afinn)

```python
!pip install afinn
```

```python
from afinn import Afinn

afinn = Afinn()

def label_senti_afinn(text):
    if isinstance(text, str):
        score = afinn.score(text)
        if score > 0:
            return "Positif"
        elif score < 0:
            return "Negatif"
        else:
            return "Netral"
    return "Netral"
```

```python
LPDP_clean['label_senti_afinn'] = LPDP_clean['english_translation'].apply(label_senti_afinn)
```
